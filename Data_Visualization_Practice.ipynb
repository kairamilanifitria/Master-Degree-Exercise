```python
# import any library
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
```

```python
# generate random data
data= pd.Series(np.random.randn(1000))

# histogram
plt.figure(figsize=(8,6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='r')

# normal distribution curve
mu, std=data.mean(), data.std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

# customize plot
plt.title('HISTROGRAM OF NORMAL DISTRIBUTION')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# show plot
plt.show()
```

```python
# boxplot

#generate random data
data=pd.DataFrame(np.random.rand(10,5),columns=['A','B','C','D','E'])
# plot
plt.figure(figsize=(5,2))
data.boxplot()

# customize boxplot
plt.title('Boxplot with Random Data')
plt.ylabel('VALUE')

# show plot
plt.show()
```

```python
# bar chart

# generate random data
data={'Category':['A','B','C','D','E'], 'Values':[23, 87, 52, 100, 78]}

# convert to dataframe
df = pd.DataFrame(data)

# plot to bar chart
plt.figure(figsize=(8,6))
plt.bar(df['Category'], df['Values'], color='skyblue')

# customize plot
plt.title('BAR CHART')
plt.xlabel('CATEGORY')
plt.ylabel('VALUES')

# show plot
plt.show()
```

```python
# Line Chart

# pandas dataframe
data = {'Years' : [2018, 2019, 2020,2021,2022, 2023,2024],
        'Sales':[1000, 1200, 1500, 1800, 2000, 2100, 2300]}
df=pd.DataFrame(data)

#Plot Line Chart
plt.figure(figsize=(5,3))
plt.plot(df['Years'],df['Sales'],marker='o',color='b',linestyle='-')

#Customize Plot
plt.title('Sales Over Years')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.grid(True)

#Show Plot
plt.show()
```

```python
# Pie Chart

# data
PenjualanEsKrim={'IceCream':['Strawberry','Vanilla', 'Cocolato','Manggo','Orange'],
                 'Value' : [30,20,15,10,25]}

# convert to pandas dataframe
df=pd.DataFrame(PenjualanEsKrim)

# plotting pie chart
plt.figure(figsize=(8,6))
plt.pie(df['Value'], labels=df['IceCream'], autopct='%1.f%%', startangle=140)

#Customize Plot
plt.title('Pie Chart of Ice Cream')

#Show Plot
plt.show()

plt.savefig('pie_chart.png')
```

```python
#SCATTER PLOT

#generate random data
np.random.seed(0)
data=pd.DataFrame({'x':np.random.randn(100), 'y':np.random.randn(100)})

#Plot Scatter
plt.figure(figsize=(10,10))
plt.scatter(data['x'],data['y'],color='r',alpha=0.6)

#Customize Plot
plt.title('SCATTER PLOT')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

#Show Plot
plt.show()
```

```python
# histogram

# generate data random
data= pd.Series(np.random.randn(1000))

# plot histogram with seaborn
plt.figure(figsize=(8,6))
sns.histplot(data, bins=30, kde=True, color='b', alpha=0.6)

#Customize Plot
plt.title('HISTOGRAM WITH SEABORN')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

#Show Plot
plt.show()
```

```python
#Boxplot

#Generate random Data
np.random.seed(0)
data=pd.DataFrame({'Category':np.random.choice(['A','B','C'], size=100),'Value':np.random.randn(100)})

#plot boxplot
plt.figure(figsize=(5,3))
sns.boxplot(x='Category',y='Value',data=data)

#customize plot
plt.title('BOXPLOT')
plt.xlabel('CATEGORY')
plt.ylabel('VALUE')

#Show Plot
plt.show()
```

```python
# Bar Chart

# generate sample data
data=pd.DataFrame({'Category':['A', 'B', 'C', 'D', 'E'],
                   'Value':[10, 20, 15, 25, 35]})

# plot bar chart
plt.figure(figsize=(8,6))
sns.barplot(x='Category', y='Value', data=data)

#customize plot
plt.title('Bar Chart')
plt.xlabel('CATEGORY')
plt.ylabel('VALUE')

#Show Plot
plt.show()
```

```python
# LINE PLOT

# generate random time series data
np.random.seed(0)
dates=pd.date_range('2022-01-01', periods=100)
data=pd.DataFrame({'Date': dates,
                   'Value':np.random.randn(100).cumsum()})

# plot line with seaborn
plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='Value', data=data)

#customize plot
plt.title('Line Chart')
plt.xlabel('Date')
plt.ylabel('Value')

#Show Plot
plt.show()
```

```python
# PIE CHART
data=pd.Series(['A', 'A', 'B', 'C', 'A', 'B', 'C', 'C', 'C', 'C'])

# Category Frequency
counts=data.value_counts()

# plot
plt.figure(figsize=(8,6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Category Frequencies')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# PIE CHART
data=pd.Series(['A', 'A', 'B', 'C', 'A', 'B', 'C', 'C', 'C', 'C'])

# Category Frequency
counts=data.value_counts()

# plot
plt.figure(figsize=(8,6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 8})  # Adjust fontsize here
plt.title('Pie Chart of Category Frequencies', fontsize=16)  # Adjust fontsize here
plt.show()
```

```python
# SCATTER PLOT

# Generate random data
np.random.seed(0)
data=pd.DataFrame({'x': np.random.randn(100), 'y': np.random.randn(100)})

# plot scatter with seaborn
plt.figure(figsize=(8,6))
sns.scatterplot(data=data, x='x', y='y')

#customize plot
plt.title('Scatter Plot with Seaborn')
plt.xlabel('X')
plt.ylabel('Y')

#Show Plot
plt.show()
```

```python
# Generate data using NumPy
np.random.seed(0)
n_data_points = 1000
data = {
    'id' : range(1, n_data_points + 1),
    'tinggi_badan' : np.random.normal(loc=170, scale=10, size=n_data_points)
}

# create DataFrame using pandas
df=pd.DataFrame(data)

# histogram using plotly
histogram = go.Histogram(x=df['tinggi_badan'],
                         marker_color='skyblue',
                         xbins=dict(start=np.min(df['tinggi_badan']),
                                    end=np.max(df['tinggi_badan']),
                                    size=10))
layout = go.Layout(title='Histogram Tinggi Badan',
                   xaxis=dict(title='Tinggi Badan'),
                   yaxis=dict(title='Frekuensi'))
fig = go.Figure(data=[histogram], layout=layout)
fig.show()
```

```python
# DataFrame
data={'Nama':['John', 'Anna', 'Peter', 'Linda'],
      'Usia': [28, 35, 40, 32]}
df=pd.DataFrame(data)

# bar chart with plotly
fig=go.Figure([go.Bar(x=df['Nama'], y=df['Usia'])])

# adding title and label
fig.update_layout(title='Diagram Batang Usia',
                  xaxis_title='Nama',
                  yaxis_title='Usia')

# show plot
fig.show()
```

```python
# DataFrame
data={
    'tahun': [2015, 2016, 2017, 2018, 2019],
    'penjualan' : [1000, 1500, 2000, 1800, 2200]
}
df = pd.DataFrame(data)

# line chart with plotly
fig = go.Figure(data=go.Scatter(x=df['tahun'], y=df['penjualan'], mode='lines+markers', name='Penjualan'))

# layout plot
fig.update_layout(title='Grafik Penjualan Tahunan',
                  xaxis_title='Tahun',
                  yaxis_title='Penjualan')

# show plot
fig.show()
```
